```python
import mlrose_hiive as mlrose
import numpy as np
import matplotlib.pyplot as plt

# Definir a função de fitness personalizada (OneMax)

# Definir a função de fitness (objetivo a minimizar)
def fitness_function(state):
    x, y = state  # Estado contém valores contínuos de x e y
    return (x - 3) ** 2 + (y + 2) ** 2  # Mínimo em (3, -2)

# Criar problema contínuo de otimização
problem = mlrose.ContinuousOpt(length=2,  # Duas variáveis (x, y)
                               fitness_fn=mlrose.CustomFitness(fitness_function), 
                               maximize=False,  # Queremos minimizar
                               min_val=-10,max_val=10)  # Restrições dos valores

# Definir os parâmetros do Hill Climbing
np.random.seed(42)
initial_state = np.random.uniform(-10, 10, size=2)

# Rodar o algoritmo Hill Climbing
best_state, best_fitness, fitness_curve  = mlrose.random_hill_climb(problem, 
                                                       max_attempts=1000, 
                                                       max_iters=1000, 
                                                       restarts=100,  # Adiciona reinicializações para evitar mínimos locais
                                                       init_state=initial_state, curve=True,
                                                       random_state=42)

fitness_curve_list = []
fitness_curve_list_2 = []
z = 1
for i in fitness_curve:
    fitness_curve_list.append(i[0])
    fitness_curve_list_2.append(z)
    z = z + 1
# Exibir os resultados
print(f"Melhor solução encontrada: {best_state}")
print(f"Melhor valor da função de fitness: {best_fitness}")

# Plotar a curva de convergência
plt.plot(fitness_curve_list_2, fitness_curve_list, label="Hill Climbing")
plt.xlabel("Iterações")
plt.ylabel("Fitness")
plt.title("Convergência do Hill Climbing")
plt.legend()
plt.show()
```

```python
import mlrose_hiive as mlrose
import numpy as np
import matplotlib.pyplot as plt

# Definir a função de fitness (objetivo a minimizar)
def fitness_function(state):
    x, y = state  # Estado contém valores contínuos de x e y
    return (x - 3) ** 2 + (y + 2) ** 2  # Mínimo em (3, -2)

# Criar problema contínuo de otimização
problem = mlrose.ContinuousOpt(length=2,  # Duas variáveis (x, y)
                               fitness_fn=mlrose.CustomFitness(fitness_function), 
                               maximize=False,  # Queremos minimizar
                               min_val=-10,max_val=10)  # Restrições dos valores

# Configurar os parâmetros do Algoritmo Genético
population_size = 500  # Número de indivíduos na população
mutation_prob = 0.1    # Probabilidade de mutação
max_attempts = 50      # Tentativas sem melhoria antes de parar
max_iters = 1000       # Número máximo de iterações

# Rodar o Algoritmo Genético
best_state, best_fitness, fitness_curve = mlrose.genetic_alg(problem, 
                                                             pop_size=population_size, 
                                                             mutation_prob=mutation_prob, 
                                                             max_attempts=max_attempts, 
                                                             max_iters=max_iters, 
                                                             curve=True, 
                                                             random_state=42)

fitness_curve_list = []
fitness_curve_list_2 = []
z = 1
for i in fitness_curve:
    fitness_curve_list.append(i[0])
    fitness_curve_list_2.append(z)
    z = z + 1
# Exibir resultados
print(f"Melhor solução encontrada: {best_state}")
print(f"Melhor valor da função de fitness: {best_fitness}")

# Plotar a curva de convergência
plt.plot(fitness_curve_list_2, fitness_curve_list, label="Genetic Algorithm")
plt.xlabel("Iterações")
plt.ylabel("Fitness")
plt.title("Convergência do Algoritmo Genético")
plt.legend()
plt.show()
```
