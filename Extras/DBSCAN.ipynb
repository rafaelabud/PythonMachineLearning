```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
import pandas as pd
```

```python
iris = datasets.load_iris()
X = iris.data[:,[2,3]]
clustering = DBSCAN(eps=0.32, min_samples=20).fit(X)
X = pd.DataFrame(X)
Base = pd.DataFrame(clustering.labels_)
Base = pd.concat([Base,X], axis =1)
Base.columns = ['Label', 'X','Y']
plt.scatter(Base['X'],Base['Y'], c=Base['Label'])
```
