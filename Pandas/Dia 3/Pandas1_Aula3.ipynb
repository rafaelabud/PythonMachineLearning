```python
#Modificando todos os nomes das colunas
df.columns = ["idade", "job", "estado civil", "educacao", "default","balance","housing", "emprestimo", 
              "contato", "dia", "mes", "duracao", "campanha", "pdays", "previous", "poutcame"]
```

```python
#Modificando apenas colunas especificas
df.rename(columns={"idade": "Idade", "job": "Profissao"}, inplace=True)
#df.drop(['Rafael'], axis = 1, inplace=True) ou df = df.drop(['Rafael'], axis = 1)
#lista = [0,3,4]
#df.drop(lista, axis = 0)
```

```python
#Manipulando o dataframe -> Info - Verifica os tipos das variáveis e se há valores nulos/faltantes
df.isnull().sum()
#df.info()
```

```python
#Manipulando o dataframe -> Describe - Exibe as estatísticas básicas dos dados numérico
df.describe()
#mediana = 7 porque separa 5,5,6,7 de 8,9,10,60 
#1_quartil = 6 porque separa 5, 5 de 6,7,8,9,10,60 
#3_quartil = 9.5 porque separa 5, 5 ,6,7,8,9 de 10,60 
idades = np.array([5,5,7,6,8,9,10,60])
df_idades = pd.DataFrame(idades)
df_idades.describe()
#idades.mean()
#Média = 13.75
#Mediana 7
```

```python
#Describe com variáveis categóricas
df.describe(include=["O"])
```
