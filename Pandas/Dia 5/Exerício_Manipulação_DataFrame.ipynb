```python
df_compradores_nomes_carros = pd.merge(df_compradores_carros,carros,left_on=['Id_carro'], right_on=['Id'], how='inner')
```

```python
df_compradores_nomes_carros = df_compradores_nomes_carros[['Id_comprador','Id_carro','Quantidade','make']]
df_compradores_nomes_carros['Nome'] = df_compradores_nomes_carros.apply(define_nomes_carros, axis = 1)
```

```python
df_compradores_nomes_carros = df_compradores_nomes_carros[['Id_comprador','Nome']]
```

```python
def agrupa(df):
    df_ = df['Nome']
    lista_linhas = []
    lista_colunas= []
    separa = df_.split('_')
    lista_linhas.append(separa[1])
    lista_colunas.append(separa[0])
    df_return = pd.DataFrame([lista_linhas])
    df_return.columns = lista_colunas
    df_return['Id_comprador'] = df['Id_comprador']
    return df_return
```

```python
lista_df = []
lista = df_compradores_nomes_carros.apply(agrupa, axis = 1)
for l in lista:
    lista_df.append(l)
```

```python
df_geral = pd.concat(lista_df)
df_geral = df_geral.fillna(0).reset_index(drop=True)
```

```python
for c in df_geral.columns:
    df_geral[c] = df_geral[c].astype(int)
```

```python
df_geral_agg = df_geral.groupby(['Id_comprador']).sum()
df_geral_agg = df_geral_agg.reset_index()
```
