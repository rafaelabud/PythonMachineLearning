```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
carros = pd.read_csv('carros.csv')
```

```python
#Hipótese 1 - Falsa, pois carros turbo não tendem a ter riskiness = -2
carros[carros.aspiration=='turbo']['riskiness'].value_counts().plot(kind='bar')
```

```python
#Hipótese 2 Carros com maior consumo de combustível na cidade tendem a ter menores riskiness - Falsa
carros.groupby(['riskiness'])['city mpg'].mean()
```

```python
sns.boxplot(x='riskiness', y='city mpg', data=carros)
```

```python
riskiness = carros['riskiness'].unique()
for r in riskiness:
    display(r)
    df = carros[carros.riskiness==r]['city mpg']
    df.plot(kind='hist')
    plt.show()
```

```python
#hipótese 3 - Carros com 4 portas são mais compridos do que aqueles com 2.
#Inicialmente, verdadeira
carros.groupby(['doors'])['length'].mean()
```

```python
import numpy as np
A = np.array([32,48,52,66])
B = np.array([12,3,2,400])
display(A.mean())
display(B.mean())
df_A = pd.DataFrame([A]).T
df_B = pd.DataFrame([B]).T
display(df_A.median())
display(df_B.median())
```

```python
sns.boxplot(x='doors', y='length', data=carros)
```
