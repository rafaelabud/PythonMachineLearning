```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1. Crie 2 arrays com as listas [-3,2,5,-2,-1,0,1,2,2,5,3] e [-27,-15.62,-8,-1,0,1,8,15,62,27]
x = np.array([-3,2,5,-2,-1,0,1,2,2,5,3])
y = np.array([-27,-15,62,-8,-1,0,1,8,15,62,27])
```

```python
#2. Use a função plot e faça um gráfico de x e y
plt.plot(x,y)
#3. Use a função xlabel para adicionar o nome "Valor de X" ao eixo x. Deixe com fonte tamanho 15
plt.xlabel('Valor de X', fontsize=15)
#4. Use a função ylabel para adicionar o nome "Valor de Y" ao eixo y. Deixe com fonte tamanho 15
plt.xlabel('Valor de X', fontsize=15)
#5. Use a função legend para adicionar o nome "Função cúbica" ao gráfico. Deixe com fonte tamanho 20
plt.legend(['Nome do gráfico'])
#6. Rode novamente seu gráfico, mas mude agora a linha para a cor magenta
plt.plot(x,y, c='red')
```
