```python
import matplotlib.pyplot as plt
from sklearn import datasets
import pandas as pd
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
def muda(valor):
    if valor=="Iris-setosa":
        return 0
    if valor=="Iris-versicolor":
        return 1
    if valor=="Iris-virginica":
        return 2
    
y.target.value_counts()
i2 = pd.read_csv('iris.data', header=None)
i2[4] = i2[4].apply(muda)
```

```python
iris = datasets.load_iris()
X = i2[[0,1,2,3]]
y = i2[4]
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=0)
X = pd.DataFrame(X_train)
X.columns = ['Petal length','Petal width','Sepal length','Sepal width']
y = pd.DataFrame(y)
y.columns = ['target']
df = pd.concat([X,y], axis = 1)
```

```python
g = sns.boxplot(y='Petal length', x='target', data=df)
g.set_xticklabels(g.get_xticklabels(),rotation=90)
plt.show()
g = sns.boxplot(y='Petal width', x='target', data=df)
g.set_xticklabels(g.get_xticklabels(),rotation=90)
plt.show()
g = sns.boxplot(y='Sepal length', x='target', data=df)
g.set_xticklabels(g.get_xticklabels(),rotation=90)
plt.show()
g = sns.boxplot(y='Sepal width', x='target', data=df)
g.set_xticklabels(g.get_xticklabels(),rotation=90)
plt.show()
```

```python
#Cria a árvore de decisão e calcula a curácia
from sklearn import tree
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=0)

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test,y_test)
from sklearn.tree import export_graphviz
export_graphviz(clf, out_file='iris2.dot',feature_names=X.columns, rounded=True, filled=True)
import pydot
(graph,) = pydot.graph_from_dot_file('iris2.dot')
graph.write_png('iris2.png')
```
