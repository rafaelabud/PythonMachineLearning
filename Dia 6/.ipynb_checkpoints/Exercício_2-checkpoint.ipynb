```python
#1. Importe a biblioteca Seaborn
import seaborn as sns
```

```python
#2. Use a função load_dataset() e coloque o dataset exercise num dataframe chamado exercícios
exercícios = sns.load_dataset('exercise')
```

```python
#3. Use a função set_style e mude para 'dark'
sns.set_style('dark')
```

```python
# Faça o histograma da variável pulse usando a função distplot e o argumento hist=False
sns.distplot(exercícios['pulse'],kde=False)
```

```python
# Agora faça a distribuição de densidade da mesma variável, mas utilizando o argumento hist=False
sns.distplot(exercícios['pulse'],hist=False)
```

```python
# Use a função swarmplot() e analias e relação entre as variáveis kind(x) e pulse(y)
sns.swarmplot(x='kind', y='pulse', data=exercícios)
```
