```python
#Matplotlib Primeiros passos
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
#y = np.array([2,4,6,8,10])
y = 2*x
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, marker='*')
```

```python
plt.plot(x,y, marker='*', c='red')
```

```python
plt.plot(x,y, marker='*', c='red', ds='steps', ls='--')
```

```python
plt.plot(x,y, marker='*', c='red', ds='steps', ls='--', mec='blue')
```

```python
plt.plot(x,y)
plt.xlabel('Valores de X', fontsize=20)
plt.ylabel('Valores de Y', fontsize=15)
plt.title('Nome do grfico', c='purple')
plt.xticks(fontsize=15, c='red')
plt.yticks(fontsize=15, c='green')
```

```python
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)
z = np.cos(x)
```

```python
plt.plot(x,y)
plt.plot(x,z)
```

```python
plt.plot(x,y,label='seno')
plt.plot(x,z,label='cosseno')
plt.plot(x,np.sqrt(x),label='raiz quadrada')
```

```python
plt.subplot(2,1,1)
plt.plot(x,y)
plt.title('Subplots')
plt.subplot(2,1,2)
plt.plot(x,z)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('Subplots')
plt.subplot(1,2,2)
plt.plot(x,z)
```
